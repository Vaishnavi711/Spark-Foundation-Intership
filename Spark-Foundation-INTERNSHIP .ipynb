```python
se.heatmap(dataset.corr(),cmap = "YlGnBu",annot = True)
```

```python
x, y = plt.subplots(nrows = 2 , ncols = 2 , figsize=(11,7))
se.countplot(dataset['Ship Mode'],ax = y[0][0])
se.countplot(dataset['Segment'],ax = y[0][1])
se.countplot(dataset['Region'],ax = y[1][0])
se.countplot(dataset['Category'],ax = y[1][1])
y[0][0].set_title("Ship Mode",fontsize = 20)
y[0][1].set_title("Segment",fontsize = 20)
y[1][0].set_title("Region",fontsize = 20)
y[1][1].set_title("Category",fontsize = 20)
plt.tight_layout()
```

```python
x, y  = plt.subplots(figsize=(10,6))
se.countplot(dataset['Sub-Category'],x=x)
plt.title("Sub-Category",fontsize=25)
plt.xticks(rotation=90)
```

```python
x, y  = plt.subplots(figsize=(10,6))
se.countplot(dataset['Category'],x=x)
plt.title("Category",fontsize=25)
plt.xticks(rotation=90)
```

```python
x1 = dataset['State'].value_counts()
x1 = x1.head(10)
x1.plot(kind = "bar",figsize=(10,6))
plt.ylabel("Numbers")
plt.xlabel("State")
plt.title("States with Top Sales")
```

```python
x1 = dataset['City'].value_counts()
x1 = x1.head(10)
x1.plot(kind = "bar",figsize=(10,6))
plt.ylabel("Numbers")
plt.xlabel("Cities")
plt.title("City with Top Sales")
```

```python
p = dataset.groupby(['Sub-Category'])[['Profit','Discount','Sales']].mean()
p = p.sort_values('Profit')
p[['Profit']].plot(kind="bar",figsize=(10,8))
plt.ylabel("Profit")
plt.show()
```

```python
p = dataset.groupby(['Sub-Category'])[['Profit','Discount','Sales']].mean()
p = p.sort_values('Sales')
p[['Sales']].plot(kind="bar",figsize=(10,8))
plt.ylabel("Sales")
plt.show()
```

```python
p = dataset.groupby(['Sub-Category'])[['Profit','Discount','Sales']].mean()
p = p.sort_values('Discount')
p[['Discount']].plot(kind="bar",figsize=(10,8))
plt.ylabel("Discount")
plt.show()
```

```python
p = dataset.groupby(['State'])[['Profit','Discount','Sales']].mean()
p = p.sort_values('Profit')
p[['Profit']].plot(kind="bar",figsize=(10,8),color = "Blue")
plt.ylabel("Profit")
plt.show()
```

```python
p = dataset.groupby(['State'])[['Profit','Discount','Sales']].mean()
p = p.sort_values('Discount')
p[['Discount']].plot(kind="bar",figsize=(10,8),color = 'gray')
plt.ylabel("Discount")
plt.show()
```

```python
p = dataset.groupby(['State'])[['Profit','Discount','Sales']].mean()
p = p.sort_values('Sales')
p[['Sales']].plot(kind="bar",figsize=(10,8),color = 'orange')
plt.ylabel("Sales")
plt.show()
```

```python
x,y = plt.subplots(nrows = 2, ncols = 2, figsize = (10,10))
se.distplot(dataset['Profit'],ax = y[0][0],color = "Blue")
se.distplot(dataset['Discount'],ax = y[0][1],color = "Green")
se.distplot(dataset['Sales'],ax = y[1][0],color = "purple")
se.distplot(dataset['Quantity'],ax = y[1][1],color = "Black")
y[0][0].set_title("Profit",fontsize = 20)
y[0][1].set_title("Discount",fontsize = 20)
y[1][0].set_title("Sales",fontsize = 20)
y[1][1].set_title("Quantity",fontsize = 20)
plt.tight_layout
```
